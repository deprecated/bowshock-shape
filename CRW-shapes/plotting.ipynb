# Plotting and computing conic information

```python
import numpy as np
import json
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

Load the json file

```python
arcdata=json.load(open("inner-isotropic-beta-0.01-m-3.0-arcdata.json"))
```

Start with $i=0^{\circ}$

```python
x = np.array(arcdata["inner"]["0.0"]["x"])
y = np.array(arcdata["inner"]["0.0"]["y"])
xc= arcdata["inner"]["0.0"]["xc"]
yc= arcdata["inner"]["0.0"]["yc"]
Rc= arcdata["inner"]["0.0"]["Rc"]
tp=np.arctan2(y,x)
R0= np.max(x)
thinf = np.array([15,30,45])
col = ["r","g","b"]
```

$(x,y)$ coordinates for conics and plot

```python
for t,c in zip(thinf,col):
    a=Rc/np.tan(np.radians(t))**2
    b=Rc/np.tan(np.radians(t))
    xe0 = a-R0
    xh0 = a+R0
    xe = a*np.cos(np.linspace(-np.pi,np.pi))-xe0
    xfe = -xe0 + np.sqrt(a**2-b**2)
    ye = b*np.sin(np.linspace(-np.pi,np.pi))
    xh = -(a*np.cosh(np.linspace(-np.pi,np.pi))-xh0)
    xfh = -xh0 + np.sqrt(a**2+b**2)
    yh = b*np.sinh(np.linspace(-np.pi,np.pi))
    plt.plot(xe,ye,c+"-",label=r"$\theta_c={}$".format(t))
    plt.plot(xh,yh,c+"-",label=r"$\theta_c=-{}$".format(t))
    plt.plot(xfe,0,c+"D")
    plt.plot(xfh,0,c+"*")
plt.plot(x,y,"k-",lw=2,label="shock")
plt.grid()
plt.plot(xc+Rc*np.cos(np.linspace(-np.pi,np.pi)),yc+Rc*np.sin(np.linspace(-np.pi,np.pi)),
         "k--",lw=1.1,label="circle least square fit")
plt.plot(xc,yc,"x",lw=2)
plt.legend(loc="best",prop=dict(size="x-small"))
plt.axis("equal")
plt.xlim(-0.9,0.1)
plt.ylim(-0.5,0.5);
```

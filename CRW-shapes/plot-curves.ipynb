# A' vs q' curves for CRW bowshocks with circular fitting

```python
import json
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

Load the json files

```python
json_pattern = "inner-rc-r0-???-arcdata.json"
lw = dict(isotropic=1,proplyd=2)
opacity = dict(isotropic=0.3,proplyd=0.7)
colors = ["#6629B1","#3030BC","#17B00B","#E3041F","#E3D104","#39A49B","#E3610A","#000000"]
innertype = ["isotropic","proplyd"]
```

```python
g = plt.figure();
for n,f in enumerate(glob.glob(json_pattern)):
    data = json.load(open(f))
    ax = g.add_subplot(2,2,n)
    for inn in innertype:
        model = data[inn]
        for b,c in zip(model.keys(),colors):
            inclinations = np.array(model[b]["inc"])
            R0 = np.array(model[b]["R0"])
            Rc = np.array(model[b]["Rc"])
            every15 = np.zeros(R0.shape,dtype=bool)
            for i15 in 0.0,15.0,30.0,45.0,60.0:
                iclosest = np.argmin(np.abs(inclinations-i15))
                every15[iclosest] = True
            label = r"$\beta={}$".format(b) if inn == "proplyd" else ""
            A = Rc/R0
            ax.plot(R0[inclinations<60],A[inclinations<60],color=c,linestyle="-",
                    lw=lw[inn],alpha=opacity[inn],label=label)
            ax.plot(R0[every15],A[every15],color=c,marker=".",alpha=opacity[inn])
    ax.legend(loc="best",ncol=2,prop=dict(size="x-small"))
#    ax.set_ylim(0.0,3)
#    ax.set_xlim(0.0,0.5)
    ax.set_title(data["info"]["Mach number"])
g.set_size_inches(10,10)
```

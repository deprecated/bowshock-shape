# A' vs q' reloaded

Graph reloaded with the new observations

```python
import numpy as np
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

First, load the observational data

```python
D=[]
R0o = []
Rco = []
Objects = ["LV2","LV3","LV4","LV5","168-328"]
path = "../../Dropbox/ProplydMIR/Jorge_prop/PC-correct/3f/"
for p in Objects:
    data = json.load(open(path+p+"-arcdata.json"))
    D.append(data["star"]["D"]) 
    R0o.append(data["outer"]["R0"])
    Rco.append(data["outer"]["Rc"])
Ap = np.array(Rco)/np.array(R0o)
```

Then, load the theoretical curves

```python
CRWdata=json.load(open("../rc-r0-fancy.json"))
colors = "bgrmkcy"
lw = dict(isotropic=1, proplyd=2)
opacity = dict(isotropic=0.3, proplyd=0.7)
for inn, modeldata in CRWdata.items():
    for b, col in zip(modeldata, colors[:len(modeldata)]):
        inclinations = np.array(modeldata[b]["inc"])
        R0 = np.array(modeldata[b]["R0'"])
        Rc = np.array(modeldata[b]["Rc"])
        # Mask to select inclinations close to multiples of 15 degrees
        every15 = np.zeros(R0.shape, dtype=bool)
        for thisinc in 0.0, 15.0, 30.0, 45.0, 60.0, 75.0:
            iclosest = np.argmin(np.abs(np.degrees(inclinations) - thisinc))
            every15[iclosest] = True
        label = r'$\beta={:.3f}$'.format(float(b)) if inn == "proplyd" else ""
        Y = Rc/R0
        # First, plot a line with all the inclinations
        plt.plot(R0, Y, '-', linewidth=lw[inn], c=col,
                 label=label, alpha=opacity[inn])
        # Second, add symbols every 15 degrees
        plt.plot(R0[every15], Y[every15], '.', c=col, alpha=opacity[inn])
        plt.plot(np.array(R0o)/np.array(D),Ap,"bo")
plt.xlim(0.0, 0.5)
plt.ylim(0.0, 5)
plt.xlabel(r"$q' = (r_0' / D')$", fontsize='x-large') 
plt.ylabel(r"$A' = R_{\mathrm{c}}' / r_0'$", fontsize='x-large') 
plt.legend(ncol=3)
plt.gcf().set_size_inches(10, 8)
```

# Testing Hyperbola and ellipse functions

Parametric equations for conics:
\begin{align}
x = a C(t) - x_0 \\
y = b S (t)
\end{align}

Where:
\begin{align}
C(t) = \cos t ~\mathrm{if~ ellipse} \\
C(t) = \cosh t ~\mathrm{if~ hyperbola} \\
S(t) = \sin t ~\mathrm{if~ ellipse} \\
S(t) = \sinh t ~\mathrm{if~ hyperbola}\\
R_c = \frac{b^2}{a}\\
\tan\theta_c = \frac{b}{a}\\
\end{align}
$a$ takes (-) sign for hyperbolae


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
t = np.linspace(-np.pi,np.pi)
beta = 0.01
R0 = np.sqrt(beta)/(1+np.sqrt(beta))
thce = 30
thch = -30
Rc = 1.2*R0
ae = Rc/np.tan(np.radians(thce))**2
ah = Rc/np.tan(np.radians(thch))**2
be = Rc/np.tan(np.radians(thce))
bh = Rc/np.tan(np.radians(thch))
x0e = ae - R0
x0h = -ah-R0
R90e = be*np.sqrt(1-(R0-ae)**2/ae**2)
R90h = -bh*np.sqrt((R0+ah)**2/ah**2-1)
sns.set_style("whitegrid")
```

```python
xe,ye = ae*np.cos(t) -x0e,be*np.sin(t)
boxstyle = {"facecolor": "white", "alpha": 0.8, "edgecolor": "none"}
fig = plt.figure()
plt.plot(xe,ye)
plt.plot([0,R0],[0,0],"k-",lw=2)
plt.plot([-x0e,-x0e+ae],[0,0],"k--")
plt.plot([-x0e,-x0e],[0,be],"k--")
plt.plot(Rc*np.cos(np.linspace(-np.pi,np.pi))-Rc+R0,Rc*np.sin(np.linspace(-np.pi,np.pi)))
m_as = -be/(ae)
as_line = m_as*np.linspace(-0.5,0.5) +m_as*x0e+be
plt.plot(np.linspace(-0.5,0.5),as_line, "k:")
plt.plot([R0-Rc,R0-Rc + Rc*np.cos(np.radians(215))],[0,Rc*np.sin(np.radians(215))],"r--")
plt.plot(0.05*np.cos(np.linspace(np.pi,np.pi-np.radians(thce)))+
         R0,0.05*np.sin(np.linspace(np.pi,np.pi-np.radians(thce))),"k-",lw=0.9)
plt.plot([R0-Rc],[0],"ko",markersize=5)
plt.plot([0,0],[0,R90e],"k-",lw=2)
plt.plot([0.0],[0.0],"r*",markersize=10)
plt.text(-0.19, 0, r"$a$", va='center', fontsize=15, bbox=boxstyle)
plt.text(-x0e, 0.5*be, r"$b$", va='center', ha='center', fontsize=15,bbox=boxstyle)
plt.text(0.01, 0.0, r"$R_0$", va='center', fontsize=15,bbox=boxstyle)
plt.text(-0.07,-0.04,r"$R_c$",fontsize=15,bbox=boxstyle)
plt.text(0.047,0.005,r"$\theta_c$", fontsize=15)
plt.text(0.0, 0.07,r"$R_{90}$", ha='center', fontsize=15,bbox=boxstyle)
plt.xlim(-0.28,0.12)
plt.ylim(-0.25,0.25)
ax = plt.gca()
ax.set_aspect("equal",adjustable="box")
ax.set_xticks([-0.2, -0.1, 0.0, 0.1])
plt.xlabel(r"$x/D$",fontsize=18)
plt.ylabel(r"$y/D$",fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=14)
fig.set_size_inches(6, 6)
fig.tight_layout()
fig.savefig("../papers/Paper1/figs/ellipse_py.pdf");
```

```python
xh,yh = -ah*np.cosh(t) -x0h,bh*np.sinh(t)
fig = plt.figure()
plt.plot(xh,yh)
plt.plot([0,R0],[0,0],"k-")
plt.plot([-x0h,-x0h-ah],[0,0],"k--")
plt.plot([R0,R0],[0,-bh],"k--")
plt.plot(Rc*np.cos(np.linspace(-np.pi,np.pi))-Rc+R0,Rc*np.sin(np.linspace(-np.pi,np.pi)))
m_as = bh/(ah)
as_line = m_as*np.linspace(-0.5,0.5) -m_as*R0-bh
plt.plot(np.linspace(-0.5,0.5),as_line, "k:")
plt.plot([R0-Rc,R0-Rc + Rc*np.cos(np.radians(215))],[0,Rc*np.sin(np.radians(215))],"r--")
plt.plot(0.05*np.cos(np.linspace(np.pi,np.pi-np.radians(thch)))+ah+R0
         ,0.05*np.sin(np.linspace(np.pi,np.pi+np.radians(thch))),"k-",lw=0.9)
plt.plot([0,0],[0,R90h],"k-",lw=2)
plt.plot([R0-Rc],[0],"ko",markersize=5)
plt.plot([0.0],[0.0],"r*",markersize=10)
plt.text(0.21, 0, r"$a$", va='center', fontsize=15,bbox=boxstyle)
plt.text(R0, -0.5*bh, r"$b$", ha='center', va='center', fontsize=15,bbox=boxstyle)
plt.text(0.5*R0, 0.0, r"$R_0$", ha='center', va='center', fontsize=15,bbox=boxstyle)
plt.text(-0.07,-0.04,r"$R_c$",fontsize=15,bbox=boxstyle)
plt.text(0.31,0.01,r"$-\theta_c$",fontsize=15)
plt.text(0.0, 0.5*R90h, r"$R_{90}$", ha='center', va='center', fontsize=15,bbox=boxstyle)
plt.xlim(-0.19,0.45)
plt.ylim(-0.32,0.32)
plt.gca().set_aspect("equal",adjustable="box")
plt.xlabel(r"$x/D$",fontsize=18)
plt.ylabel(r"$y/D$",fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=14)
fig.set_size_inches(6,6)
fig.tight_layout()
plt.savefig("../papers/Paper1/figs/hyperbola_py.pdf");
```

## $R_{90}$ for ellipse

$t_{90}$ is such that:
\begin{align}
x_{90} = R_0 - a = a\cos t_{90} \\
y_{90} = R_{90} = b\sin t_{90} \\
R_{90} = b\left(1-\frac{(R_0-a)^2}{a^2}\right)^{1/2}
\end{align}

## $R_{90}$ for hyperbola

in this case:
\begin{align}
x_{90} = a\cosh t_{90} = R_0 - a\\
y_{90} = R_{90} = b\sinh t_{90} \\
R_{90} = b\left(\frac{(R_0 - a)^2}{a^2}-1\right)^{1/2}
\end{align}

## $R_{90}$ general

$R_{90} = b\left[\pm \left(1-\frac{(R_0 - a)^2}{a^2}\right)\right]^{1/2}$

(+) for ellipse
(-) for hyperbola

## Compare a proplyd and isotropic model with the same value of $A$

```python
pdata = json.load(open("CRW-shapes/proplyd-beta-0.02-arcdata.json"))['outer']['i=0.0']
idata = json.load(open("CRW-shapes/isotropic-beta-0.001-arcdata.json"))['outer']['i=0.0']
iidata = json.load(open("CRW-shapes/isotropic-beta-0.01-arcdata.json"))['outer']['i=0.0']
```

```python
pdata['Rc'], idata['Rc'], iidata['Rc']
```

```python
xp, yp = np.array(pdata['x']), np.array(pdata['y'])
xi, yi = np.array(idata['x']), np.array(idata['y'])
xii, yii = np.array(iidata['x']), np.array(iidata['y'])
```

```python
R0p = xp.max()
R0i = xi.max()
R0ii = xii.max()
xp /= R0p
yp /= R0p
xi /= R0i
yi /= R0i
xii /= R0ii
yii /= R0ii
```

```python
Rcp = pdata['Rc']/R0p
Rci = idata['Rc']/R0i
Rcii = iidata['Rc']/R0ii
Rcp, Rci, Rcii
```

```python
Rc = Rcp
th = np.linspace(0.0, np.pi, 500)
xc, yc = Rc*np.cos(th) - (Rc - 1), Rc*np.sin(th)


plot(xp, yp, "g-", alpha=0.4, lw=4)
plot(xi, yi, "g--", alpha=0.4, lw=4)
#plot(xii, yii)
plot(xc, yc, label="circle")

for theta in [24.0, 26., 28., 30., 32.]:
    xcc, ycc = conic_utils.world_hyperbola(Rh=Rc, thh=theta, 
                                           PAh=90.0, xxh=-(Rc - 1.), yyh=0.0,
                                           xmin=-3.0, xmax=0., N=50, full=False)
    plot(xcc, ycc, label="i = {:.0f}".format(theta))

th0 = 110.0
rr = np.linspace(0.0, 4.0, 2)
xx, yy = rr*np.cos(np.radians(th0)), rr*np.sin(np.radians(th0))
plot(xx, yy, "k--", lw=4, alpha=0.2)
xlim(-2.0, 1.0)
ylim(0.0, 3.0)
legend()
grid()
gcf().set_size_inches(8, 8)
```

```python
def polar(x, y, x0=0):
    return np.hypot(x-x0, y), np.arctan2(y, x-x0)

rp, thp = polar(xp, yp, x0=-(Rc - 1.))
rc, thc = polar(xc, yc, x0=-(Rc - 1.))
plot(np.degrees(thp), rp, lw=3, alpha=0.5)
plot(np.degrees(thc), rc)

for theta in [24.0, 26., 28., 30., 32.]:
    xcc, ycc = conic_utils.world_hyperbola(Rh=Rc, thh=theta, 
                                           PAh=90.0, xxh=-(Rc - 1.), yyh=0.0,
                                           xmin=-3., xmax=3., N=500, full=True)
    rcc, thcc = polar(xcc, ycc, x0=-(Rc - 1.))
    plot(np.degrees(thcc), rcc, label=str(theta))
xlim(-100., 100.0)
ylim(0.9*Rc, 1.5*Rc)
legend(loc="upper left")
xlabel('Theta around center of curvature')
ylabel('Radius from center of curvature')
gcf().set_size_inches(10, 10)
```

```python
rp, thp = polar(xp, yp, x0=0)
rc, thc = polar(xc, yc, x0=0)

th1, th2 = 0.0, 150.0
fill_between([th1, th2], [-0.03, -0.03], [0.03, 0.03], alpha=0.05)
fill_between([th1, th2], [-0.01, -0.01], [0.01, 0.01], alpha=0.1)
fill_between([th1, th2], [-0.005, -0.005], [0.005, 0.005], alpha=0.1)

err_p_c = (rp - np.interp(thp, thc, rc))/rp
plot(np.degrees(thp), err_p_c, lw=3, alpha=0.5, label="circle")

mp = (thp > np.radians(th1)) & (thp < np.radians(th2))
for theta in [24.0, 26., 28., 30., 32.]:
    xcc, ycc = conic_utils.world_hyperbola(Rh=Rc, thh=theta, 
                                           PAh=90.0, xxh=-(Rc - 1.), yyh=0.0,
                                           xmin=-3., xmax=0., N=500, full=False)
    rcc, thcc = polar(xcc, ycc, x0=0)
    m = np.isfinite(rcc) & np.isfinite(thcc) 
    m = m & (thcc > np.radians(th1)) & (thcc < np.radians(th2))
    err_p_cc = (rp[mp] - np.interp(thp[mp], thcc[m][::-1], rcc[m][::-1]))/rp[mp]
    plot(np.degrees(thp[mp]), err_p_cc, label=str(theta), lw=3, alpha=0.5)
xlim(th1, th2)
ylim(-0.1, 0.1)
legend(loc="upper left")
xlabel('Theta around star')
ylabel('Relative error, (R(exact) - R(approx)) / R(exact)')
grid()
gcf().set_size_inches(8, 8)
```

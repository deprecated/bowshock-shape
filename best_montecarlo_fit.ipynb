# Best Distribution for D'

Reproduce the class for the simulated distributions for D'

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Population(object):
    
    def __init__(self,N,Dmin=0,Dmax=800,m=2.6):
        assert (m<3)
        self.N = N
        self.m = m
        self.Dmin = Dmin
        self.Dmax = Dmax
        
        #set a set of random numbers to do a distribution of Ds
        #I think D is such that n ~ D^-m
        xi = np.random.random_sample((N,))
        x0 = Dmin/Dmax
        self.D = Dmax*((1 - x0**(3.0 - m))*xi + x0**(3.0 - m))**(1.0/(3.0 - m))
        
        #Set of mu -> [-1,1]
        xi = np.random.random_sample((N,))
        self.mu = 2*xi - 1.0
        
         # Inclination to plane of sky inc = [-90:90]
        self.inc = np.degrees(np.arcsin(self.mu))
        
        # Projected distance
        self.Dprime = self.D*np.sqrt(1.0 - self.mu**2)
        
        # Now for position angle
        xi = np.random.random_sample((N,))
        self.PA = 360.0*xi
        
        # Plane-of-sky coords (Right Ascention and Declination)
        self.Dec = self.Dprime*np.cos(np.radians(self.PA))
        self.RA = self.Dprime*np.sin(np.radians(self.PA))
```

The next steps will be:
* Obtain HA 98 Data, Ricci data and/or other else
* Obtain the best model via chi square minimization

Start with Henney & Arthur 1998 data

```python
from astropy.table import Table
```

```python
datafile = "Proplyd-Datasets/vicente-vs-HA98.dat"
data = Table.read(datafile, format='ascii')
```

```python
data
```

```python
HA_D = data["d"]
```

```python
MC_D = Population(28,Dmin = 2.1,Dmax= 150) 
```

Now, it's time to define our $\chi^2$ function

```python
def chi_sq(data,model):
    ch2 = 0
    for d,m in zip(data,model):
        ch2 += (d-m)**2/m
    return ch2
```

```python
chi2 = chi_sq(HA_D,MC_D.Dprime)
```

```python
plt.hist(HA_D,bins = 30,range = (0,28),cumulative = True,color="b",alpha=0.3,label = "Henney & Arthur 1998")
plt.hist(MC_D.Dprime,bins=30,range=(0,28),cumulative=True,color="r",alpha=0.3,label = "Montecarlo simulated data")
plt.xlabel("Projected D', arcsec")
plt.ylabel("No of proplyds with projected distance < D'")
plt.title("Cumulative Distribution of Proplyds vs Distance: Henney & Arthur Sample vs Montecarlo Random Sample")
plt.legend(loc="upper left")
```

The next step is to found a value for m that minimizes $\chi^2$

```python
M = np.linspace(0.1,2.7,500)
x2 = []
for m in M:
    model = Population(28,Dmin=2,Dmax=150,m=m)
    sq = chi_sq(HA_D,model.Dprime)
    x2.append(sq)
```

```python
plt.plot(M,x2,"b.")
#plt.xlim(0,2.5)
plt.ylim(0,5000)
plt.xlabel("m")
plt.ylabel(r"$\chi^2$")
plt.title(r"$\chi^2$ test")
```

A juzgar por el test y considerando el ruido que tiene, parece que el mejor ajuste ronda entre 2 y 2.5

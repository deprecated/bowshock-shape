# Best Distribution for D'

Reproduce the class for the simulated distributions for D'

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
class Population(object):
    
    def __init__(self,N,Dmin=0,Dmax=800,m=2.6,r0=50.0,a=1.1):
        assert (m<3)
        self.N = N
        self.m = m
        self.Dmin = Dmin
        self.Dmax = Dmax
        
        #set a set of random numbers to do a distribution of Ds
        #I think D is such that n ~ D^-m
        xi = np.random.random_sample((N,))
        x0 = Dmin/Dmax
        self.D = Dmax*((1 - x0**(3.0 - m))*xi + x0**(3.0 - m))**(1.0/(3.0 - m))
        
        #Set of mu -> [-1,1]
        xi = np.random.random_sample((N,))
        self.mu = 2*xi - 1.0
        
         # Inclination to plane of sky inc = [-90:90]
        self.inc = np.degrees(np.arcsin(self.mu))
        
        # Projected distance
        self.Dprime = self.D*np.sqrt(1.0 - self.mu**2)
        
        # Now for position angle
        xi = np.random.random_sample((N,))
        self.PA = 360.0*xi
        
        # Plane-of-sky coords (Right Ascention and Declination)
        self.Dec = self.Dprime*np.cos(np.radians(self.PA))
        self.RA = self.Dprime*np.sin(np.radians(self.PA))
        #Distribution of proplyd sizes in AU
        rmax = 6*self.D
        b = 1.0 - 1.0/(1.0+(rmax/r0)**(2*a))
        xi = np.random.random_sample((N,))
        self.rAU = r0*((1.0/(1.0-b*xi))-1.0)**(1./(2*a))
```

Now, it's time to define our $\chi^2$ function

```python
def chi_sq(data,model):
    ch2 = 0
    for d,m in zip(data,model):
        ch2 += (d-m)**2/m
    return ch2
```

```python
chi2 = chi_sq(sorted(HA_D),sorted(MC_D.Dprime))
```

```python
plt.hist(HA_D,bins = 30,range = (0,28),cumulative = True,color="b",alpha=0.3,label = "Henney & Arthur 1998")
plt.hist(MC_D.Dprime,bins=30,range=(0,28),cumulative=True,color="r",alpha=0.3,label = "Montecarlo simulated data")
plt.xlabel("Projected D', arcsec")
plt.ylabel("No of proplyds with projected distance < D'")
plt.title("Cumulative Distribution of Proplyds vs Distance: Henney & Arthur Sample vs Montecarlo Random Sample")
plt.legend(loc="upper left")
```

The next step is to found a value for m that minimizes $\chi^2$

```python
M = np.linspace(0.1,2.7,500)
x2 = []
for m in M:
    model = Population(28,Dmin=2,Dmax=150,m=m)
    sq = chi_sq(HA_D,model.Dprime)
    x2.append(sq)
```

```python
plt.plot(M,x2,"b.")
#plt.xlim(0,2.5)
plt.ylim(0,5000)
plt.xlabel("m")
plt.ylabel(r"$\chi^2$")
plt.title(r"$\chi^2$ test")
```

Repeat the same exercise inluding the recombinations in the flux term

The idea is the next: In ionization balance, the total flux of ionizing radiation is given by:
\begin{equation}
N = \int^{\infty}_{\nu_T} N(\nu) exp[-\tau(\nu,r)]~d\nu
\end{equation}

Where:
    $N(\nu) = \frac{L_*(\nu)}{4\pi r^2 h\nu}$
    
$\tau(\nu,r) = \int^{r}_{0}\alpha_H(\nu)n_H(r')~dr'$

$L_*$ is the star luminosity, $\alpha_H$ is the ionization cross section for hydrogen, $n_H$ is the density and $\tau$ is the optical depth

The total photons emmited by the star is given by:

\begin{equation}
N_{Lyc} = \int^{\infty}_{\nu_T} \frac{L_*(\nu)}{h\nu}~d\nu 
\end{equation}

For small optical depth, we can do the following:
    \begin{equation}
    N = \int^{\infty}_{\nu_T} N(\nu)(1-\tau(\nu,r))~d\nu
    \end{equation}

So:
\begin{equation}
N = \frac{1}{4\pi r^2}\int^{\infty}_{\nu_T}\frac{L_*(\nu)}{h\nu}~d\nu - \frac{1}{4\pi r^2}
\int^{\infty}_{\nu_T}\frac{L_*(\nu)\alpha_H(\nu)}{h\nu}\left(\int^{r}_{0} n_H(r')~dr'\right)d\nu
\end{equation}

Assuming constant density, we get:

\begin{equation}
N\simeq \frac{1}{4\pi r^2}\int^{\infty}_{\nu_T}\frac{L_*(\nu)}{h\nu}~d\nu -\frac{n_H}{4\pi r}\int^{\infty}_{\nu_T} \frac{L_*(\nu)\alpha_H(\nu)}{h\nu}~d\nu
\end{equation}

\begin{equation}
N \simeq \frac{N_{Lyc}}{4\pi r^2} - \frac{n_H}{4\pi r}\int^{\infty}_{\nu_T}\frac{L_*(\nu)\alpha_H(\nu)}{h\nu}~d\nu
\end{equation}

Now, the averaged cross section for hydrogen is defined as:

\begin{equation}
\bar{\alpha}_H \equiv \frac{1}{N} \int^{\infty}_{\nu_T}\frac{L_*(\nu)}{4\pi r^2 h\nu}\alpha_H(\nu)~d\nu
\end{equation}

Then:
\begin{equation}
N \simeq \frac{N_{Lyc}}{4\pi r^2}- N\bar{\alpha}_H n_H r
\end{equation}

Finally:
\begin{equation}
N \simeq \left(\frac{N_{Lyc}}{4\pi r^2}\right)\left(1-\bar{\alpha}_H n_H r\right)
\end{equation}

For a typical HII region, $\bar{\alpha}_H = 3 \times 10^{-19}~\mathrm{cm^2}$

Sounds good, except that the quantity $\bar{\alpha}_H n_H r$ is not small, so this approximation does not work (or at least it seems so)

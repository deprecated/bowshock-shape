## True $\theta$ versus projected $\theta'$ for conic sections

```python
import numpy as np

def theta(t, a, b, x0):
    return np.degrees(np.arctan2(b*np.sin(t), a*np.cos(t) - x0))

def thetac(t, a, b, x0):
    return np.degrees(np.arctan2(b*np.sin(t), a*np.cos(t)))

def thetaprime(t, inc, a, b, x0):
    numerator = b*np.sqrt(np.sin(t)**2 - ((b/a)*np.tan(inc)*np.cos(t))**2)
    denominator = np.cos(inc)*(a*np.cos(t)*(1.0 + (b*np.tan(inc)/a)**2) - x0)
    return np.degrees(np.arctan2(numerator, denominator))
```

```python
a = 1.0
b = np.tan(np.radians(30.0))
x0 = 0.8
trange = np.linspace(0.0, np.pi, 2001)
for incd in [0.0, 15.0, 30.0, 45.0, 60.0, 75.0, 89.6]:
    inc = np.radians(incd)
    plot(thetaprime(trange, inc, a, b, x0), theta(trange, a, b, x0),
         '-',
         label="i = {:.0f}".format(incd))
#    plot(thetaprime(trange, inc, a, b, x0), thetac(trange, a, b, x0), ':')
xlabel(r"$\theta'$")
ylabel(r"$\theta$")
legend(ncol=2, loc="lower right")
gcf().set_size_inches(6, 6)
```

So, for $\theta' < 60^\circ$ and $i < 60^\circ$, we need $\theta$ up to $110^\circ$.

If we wanted to go to $i = 75^\circ$ then we would need up to $\theta = 130^\circ$.

## Projected shape factors: $A'$ versus $q'/q$ for conic sections

These are the results for spheroids that I derived in my handwritten notes.  $A = R_ {\mathrm{c}} / R_ 0$ and $A'$ is the projected version of the same.  $q = R_ 0/ D$ and $q'$ is the projected version of the same.  

```python
def g(inc, A, thetac):
    return 1 + A*(f(inc, thetac) - 1.0)/np.tan(thetac)**2

def f(inc, thetac):
    return np.sqrt(1 + np.tan(thetac)**2 * np.tan(inc)**2)

def Aprime(inc, A, thetac):
    return A/(np.cos(inc)**2 * f(inc, thetac) * g(inc, A, thetac))

A = 1.6
incs = np.radians([0.0, 15.0, 30.0, 45.0, 60.0, 75.0])
incrange = np.linspace(0.0, 0.5*np.pi, 90)
for A, ls in [1.2, "-"], [1.6, "--"], [2.0, ":"]:
    for thetac, c in zip(np.radians([45.0, 28.0, 0.1]), "gbr"):
	x = g(incrange, A, thetac)
	y = Aprime(incrange, A, thetac)
	plot(x, y, c + ls,
	     label=r"$A = {:.1f}$, $\theta_{{\mathrm{{c}}}} = {:.0f}^\circ$".format(A, np.degrees(thetac)))
	x = g(incs, A, thetac)
	y = Aprime(incs, A, thetac)
	plot(x, y, c + "o")

xlim(0.0, 10.0)
ylim(0.0, 3.0)
xlabel(r"$q'/q$") 
ylabel(r"$A'$") 
legend(ncol=3)
gcf().set_size_inches(8, 6)
```

The dots are shown every $15^\circ$ up to $75^\circ$.  There is a limiting value of $A'$ for each value of $\theta_ {\mathrm{c}}$, which is the limit as $i \to 90^\circ$.  This is 1 for circles $\theta_ {\mathrm{c}} = 45^\circ$ and 2 for parabolas $\theta_ {\mathrm{c}} = 0$.

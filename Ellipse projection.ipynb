## Projected shape factors: $A'$ versus $q'/q$ for conic sections

These are the results for spheroids that I derived in my handwritten notes.  $A = R_ {\mathrm{c}} / R_ 0$ and $A'$ is the projected version of the same.  $q = R_ 0/ D$ and $q'$ is the projected version of the same.  

```python
incs = np.array([0.0, 15.0, 30.0, 45.0, 60.0, 75.0])
incrange = np.linspace(0.0, 90.0, 500)
for A, ls in [1.2, "-"], [1.6, "--"], [2.0, ":"]: #, [2.5, "-."]:
    for thetac, color in zip([60.0, 45.0, 28.0, 0.1, -15.0, -30.0], "cmbryg"):
        c = conproj_utils.Conic(A=A, th_conic=thetac)
        x = c.g(incrange)
        y = c.Aprime(incrange)
        plot(x, y, color + ls, 
             label=r"$A = {:.1f}$, $\theta_{{\mathrm{{c}}}} = {:.0f}^\circ$".format(A, thetac))
        x = c.g(incs)
        y = c.Aprime(incs)
        plot(x, y, color + "o")

xlim(0.0, 10.0)
ylim(0.0, 7.5)
xlabel(r"$q'/q = (r_0' / D') / (r_0 / D)$", fontsize='x-large') 
ylabel(r"$A' = R_{\mathrm{c}}' / r_0'$", fontsize='x-large') 
legend(ncol=3)
gcf().set_size_inches(10, 8)
gcf().savefig("projected-shapes-ellipse.pdf")
```

The dots are shown every $15^\circ$ up to $75^\circ$.  There is a limiting value of $A'$ for each value of $\theta_ {\mathrm{c}}$, which is the limit as $i \to 90^\circ$.  This is 1 for circles $\theta_ {\mathrm{c}} = 45^\circ$ and 2 for parabolas $\theta_ {\mathrm{c}} = 0$.

```python
fig, axes = subplots(2, 3, sharex=True, sharey=True)
thetac_list = [60.0, 45.0, 28.0, 0.1, -15.0, -30.0]
desc_list = ["oblate spheroid", "sphere", "prolate spheroid", 
                "paraboloid", "hyperboloid", "hyperboloid"]
for thetac, ax, desc in zip(thetac_list, axes.ravel(), desc_list):
    Alist = np.linspace(1.0, 3.0, 11)
    incs = np.array([0.0, 15.0, 30.0, 45.0, 60.0, 75.0])
    incrange = np.linspace(0.0, 90.0, 500)
    label = desc + r': $\theta_{{\mathrm{{c}}}} = {}$'.format(int(thetac))
    ax.text(1.0, 0.25, 
            label,
            fontsize='x-large'
            )
    for A in Alist:
        c = conproj_utils.Conic(A=A, th_conic=thetac)
        x = c.g(incrange)
        y = c.Aprime(incrange)
        ax.plot(x, y, '-k', alpha=0.4)
        x = c.g(incs)
        y = c.Aprime(incs)
        ax.scatter(x, y, c=incs, vmin=0.0, vmax=90.0)
        ax.grid(color='b', ls='-', alpha=0.2)
axes[0,0].set_xlim(0.0, 8.0)
axes[0,0].set_ylim(0.0, 4.0)
for ax in axes[-1,:]:
    ax.set_xlabel(r"$q'/q = (r_0' / D') / (r_0 / D)$", fontsize='x-large') 
for ax in axes[:,0]:
    ax.set_ylabel(r"$A' = R_{\mathrm{c}}' / r_0'$", fontsize='x-large') 
#legend(ncol=3)
fig.set_size_inches(12, 6)
fig.tight_layout()
fig.savefig("projected-shapes-ellipse.pdf")
```

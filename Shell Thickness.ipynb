### Approximations to the bowshock shape

For the equal momentum case, the CD is a plane, so $R = z_0 /\cos\theta$ with $R \to \infty$ as $\theta \to \frac12\pi$.  For the general case, we have conical wings with asymptotic angle $\theta_{\infty} > \frac12 \pi$. So we could try $R = z_0 /\cos a \theta$ with $a = \frac12 \pi / \theta_{\infty}$.

```python
z0 = 1.0
fitdata = {}
theta_infs = np.arange(120, 180)
b = 1.0
for theta_inf in theta_infs:
    a = 90.0/theta_inf
    theta = np.linspace(0.0, theta_inf, 100, endpoint=False)
    R = z0/np.cos(a*np.radians(theta))**b
    x = R*np.cos(np.radians(theta))
    y = R*np.sin(np.radians(theta))
    fitdata[theta_inf] = dict(a=a, theta=theta, R=R, x=x, y=y)
    plot(x, y)
xlim(-5.0, 1.0)
ylim(0.0, 6.0)
gcf().set_size_inches(8, 8)
```
